# รับค่าตัวเลขจากผู้ใช้ 2 จำนวน แล้วหาค่ามากที่สุด

```python
user1 = float(input("number"))
user2 = float(input("number"))

cat = max(user1,user2) # "max" เป็นฟังก์ชั่นในไพทอน เอาไว้หาค่าที่มากที่สุด
print(cat)
```

# รับค่าขนาดหน้าอก เพื่อท าการตรวจสอบขนาดของเสื้อโดยมีเงื่อนไขดังนี้
- ถ้าขนาดหน้าอกไม่เกิน 34 จะได้เสื้อขนาด XS
- ถ้าขนาดหน้าอกไม่เกิน 36 จะได้เสื้อขนาด S
- ถ้าขนาดหน้าอกไม่เกิน 38 จะได้เสื้อขนาด M
- ถ้าขนาดหน้าอกไม่เกิน 40 จะได้เสื้อขนาด L
# - ถ้าขนาดหน้าอกมากกว่า40 จะได้เสื้อขนาด XL

```python

user = float(input("your size number"))

if user <= 34 :
    print("XS")
elif user <= 36 :
    print("S")
elif user <= 38 :
    print("M")
elif user <= 40 :
    print("L")
else :
     print("Xl")
```
